```python
from skimage.transform import resize
from matplotlib.pyplot import imshow
from matplotlib.image import imread, imsave
from matplotlib import pyplot as plt
from glob import glob
import pandas as pd
import numpy as np
import os
import cv2
import dlib
import face_recognition
dlib.DLIB_USE_CUDA
```

```python
DATA_PATH = 'celeba-dataset/img_align_celeba'
data = glob(os.path.join(DATA_PATH, "*.jpg"))
```

```python
len(data)
```

```python
counter = 0
for file_name in data:
    
    # for every 1000 images print a progess status
    counter = counter + 1
    if counter % 1000 == 0:
        print(counter)
        
    # recognize face(s)
    img = face_recognition.load_image_file(file_name)
    face_location = face_recognition.face_locations(img,\
                                                          number_of_times_to_upsample=0)
    
    # make sure there is only one face in each image
    if len(face_location) != 1:
        print('{}--- find {:2d} at {} ---'.format(counter,\
                                                  len(face_location),\
                                                  file_name.strip().split('/')[-1]))
        continue
    
    # crop face to variable face
    for top, right, bottom, left in face_location:
        face = img[top:bottom, left:right]
    
    # resize to 64x64
    face = resize(image=face, output_shape=[64,64,3], mode='reflect', anti_aliasing=True)
    
    # save to jpg file
    imsave(os.path.join('celeba-dataset/face',\
                        file_name.strip().split('/')[-1]), face)
```
